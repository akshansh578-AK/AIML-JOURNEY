### VARIABLES

Variables are fundamental elements in programming which is used to store data that can be used and manipulated any where in the program. 

Variables are created when we assign a value to it and they do not need any prior declaration , the declaration occurs automatically when we assign a value to it.

```python
## Declaring and Assigning Variables:

Name = "AKSHANSH"
Age = 20
Height = 5.6
is_Student = True

print("NAME:",Name)
print("AGE:",Age)
print("HEIGHT",Height)
```

```python
## Naming convention
## Variable name can be anything except keyword.
## It must start with a letter or an '_' and then can contain any letter , number.
## Variable name are case Sensitive

## Valid Variable name

first_name = "AKSHANSH"
last_name = "MISHRA"
```

```python
## Invalid variable name

##9ak = 45
## Student-name = "AKSHANSH"
## $name = "AK"
```

```python
## Case sensitive :
name = "AKSHANSH"
Name = "MISHRA"
```

```python
## Understanding variable types : Python is dynamically typed , type of variable is defined at runtime.

age = 35 ## Int
height = 5.6 ## Float
name = "SHRIKANT" ## String
is_Spy = False ## Boolean

print(type(age))
```

```python
## Typecasting : Converting one data to another data type.

age = 30
age_str = str(age)

print(type(age))
print(type(age_str))
```

```python
name = "AKSH"
print(type(int(name)))
```

```python
Height = 5.9
a=int(Height)
type(a)
```

```python
## Dynamic Typing: Python allows type of variable to change as the program executes.

var = 10 
print(var,type(var))
var = "10"
print(var,type(var))
```

```python
## Taking input from user 
age = input("Enter Your Age: ")
print(age)
print(type(age))

## It always return as String

age = int(input("Enter your age: "))
print(age)
```
