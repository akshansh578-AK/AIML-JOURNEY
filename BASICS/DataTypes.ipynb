DATA TYPES

.Data types are classification of data which tell the compiler or interpreter that how the program use the data.
They tell about the the type of operation can be performed on data and the values data can take , and the amount of memory needed to store data.

```python
## Integers Example

age = 35
type(age)
```

```python
## Float Example
age = 4.5
type(age)
```

```python
## String Example 
Name = "AKSHANSH"
type(Name)
```

```python
## Boolean Example
a = True
type(a)
```

```python
a = 4*5
b = 5*4

type(a==b)
```

```python
## Common errors:

## Here we will get error because concatenation is between string and integer.
result = "Hello" + 6

## Here we will not get error because concatenation is between strings.
result = "Hello" + "6"
print(type(result))
```
