OPERATORS: Symbols used to perform different mathematical operations on different data types.

```python
## Arithmetic Operation

a = float(input("Enter The First Number: "))
b = float(input("Enter The Second Number: "))

Sum = a+b
Differnce = a-b
Multiplication = a*b
Division = a/b
Floor_Division = a//b
Modulus = a%b ## Gives remainder on division
Exponent = a**b

print(Sum)
print(Differnce)
print(Multiplication)
print(Division) ## Returns the exact quotient as a floating-point number.
print(Floor_Division) ## Returns the quotient rounded down to the nearest integer.
print(Modulus)
print(Exponent)
```

```python
## Greater than equal to >=

a = 10
b = 10

print(a>=b)
```
