```python
# Python is case sensitive.

## Like we initialise and declare variables and we give same name to both but just differentiate the words with uppercase and lowercase then both the variable name will be treated as different identifier.

# Identifier are the name we give to variables , functions, classes etc.

Age = 19
age = 20

print(Age)
print(age)
```

###Indentation

Indentation in python is used to define block of code and determine the grouping of statements which means all the statements within a block must be indented at the same level.

```python
age = 32
if age>30:
    print(age)

print(age)    
```
