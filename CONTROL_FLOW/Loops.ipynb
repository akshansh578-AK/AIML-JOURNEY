##Loops : It is an concept of programming in which a set of instructions is executed multiple times , without writing the same code again for the same.

```python
range(5) ## It gives sequence of number from zero to n-1.
```

```python
## For loop

for i in range(5):
    print(i)
```

```python
for i in range(1,7):
    print(i)
```

```python
# start → the number to begin from (inclusive).
# stop → the number to stop before (exclusive).
# step → how much to increase or decrease the value after each     iteration.
# range(start,stop,step)
range(1,10,2)
```

```python
for i in range(1,9,2):
    print(i)
```

```python
for i in range(10,1,-2):
    print(i)
```

```python
str = "AKSHANSH"

for i in str:
    print(i)
```

```python
## While Loop: Continues to execute th eblock of code as long as condititon is True.

count = 0

while count<5:
    print(count)
    count = count + 1
```

```python
## Loop Control Statement.

#Break : The break statement exits the loop prematurely(Means ended before it had to end.)

for i in range(1,8,2):
    if i==7:
        break
    print(i)
```

```python
## continue : It skips the given itteration and continues with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)

```

```python
## pass : It is a null operation and do nothing.

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested Loops : Loop inside Loop 

# formatted string literal (f-string) in Python.
# f before the quotes tells Python to replace the values inside {} with the actual variable values.
# {i} is replaced by the value of i.
# {j} is replaced by the value of j.

for i in range(4):
    for j in range(3):
        print(f"i:{i} and j:{j}")
```

```python
## Calculate the sum of first N natural numbers using a while and for loop.

# While loop

num = int(input("Enter the number: "))
sum = 0
count = 1

while count<=num:
    sum = sum + count
    count = count + 1

print("SUM:",sum)    
```

```python
num = 10 
sum = 0

for i in range(1,11):
    sum = sum + i

print("SUM:",sum)    
```

```python
## Prime number between 1 and 100.

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)    
```
